## The TimeDelta Object

````python
import pandas as pd
````

````python
aug = pd.to_datetime("2015-08-15")
aug
````

````python
sep = pd.to_datetime("2015-09-02")
sep
````

````python
# minus between datetime objects makes timeDelta
delta = sep - aug
delta
````

````python
type(delta)
````

````python
aug + delta
````

````python
sep - delta
````

````python
aug - sep
````

````python
SP500["BQE"] = SP500.index + pd.tseries.offsets.BQuarterEnd()
````

````python
SP500.head()
````

````python
SP500["TDelta"] = SP500["BQE"] - SP500.index
````

````python
SP500.head()
````

````python
SP500.info()
````

````python
SP500 = SP500.reset_index().set_index("TDelta")
````

````python
SP500
````
