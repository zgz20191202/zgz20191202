96. __Create__ a __DatetimeIndex__ starting from __May 17, 2019__ (incl.) and including the __following 9 Fridays__ (in total 10 Fridays incl. May 17, 2019)!

```python
pd.date_range("May 17, 2019",periods=9,freq="w-fri")
```
